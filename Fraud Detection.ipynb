```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data
```

```python
# In the above dataset, the author has used PCA(Principal Component Analysis method)
# which converts the present components into numericals
```

```python
credit_card_data.isnull().sum()
# isnull().sum() is used to find no. of missing values or null values
```

```python
# Dataset information
credit_card_data.info()
```

```python
# distribution of legit transaction and fraudulent transaction
credit_card_data['Class'].value_counts(sort = True,ascending = True)
# Here o - Nrml transaction
# here 1 - fraudulent transaction
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
legit
```

```python
fraud = credit_card_data[credit_card_data.Class == 1]
fraud
```

```python
legit.describe()
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Pandas groupby is used for grouping the data according to the categories and applying a function to the categories
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
legit_sample
```

```python
# Concatenating two dataframe
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the dataset into features and targets
X = new_dataset.drop(columns = 'Class',axis = 1)
```

```python
X
```

```python
y = new_dataset['Class']
y
```

```python
# Split the data into train and test data
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,stratify=y,random_state=2)
```

```python
X.shape,X_train.shape,X_test.shape,y.shape,y_train.shape,y_test.shape
```

```python
# Model Training

model = LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train,y_train)
```

```python
# Model Evaluation - Accuracy Score
# Accuracy score on training data

X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict,y_train)
training_data_accuracy
```

```python
pred = model.predict(X_test)
test_data_accuracy = accuracy_score(pred,y_test)
test_data_accuracy
```
